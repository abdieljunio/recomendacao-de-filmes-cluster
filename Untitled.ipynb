```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Definir os nomes das colunas para os dados de avaliações
r_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
# Carregar o arquivo u.data
ratings = pd.read_csv(
    'ml-100k/u.data', 
    sep='\t', 
    names=r_cols, 
    encoding='latin-1'
)

# Definir os nomes das colunas para os dados dos filmes
# (Carregando todas as colunas, incluindo gêneros)
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'IMDb_URL', 
          'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 
          'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 
          'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
# Carregar o arquivo u.item
movies = pd.read_csv(
    'ml-100k/u.item', +
    sep='|', 
    names=m_cols, 
    encoding='latin-1'
)

# Guardar os nomes das colunas de gênero para usar depois
genre_cols = m_cols[5:] # Pega da coluna 'unknown' até 'Western'

print("--- Amostra de Avaliações (ratings) ---")
print(ratings.head())
print("\n--- Amostra de Filmes (movies) ---")
print(movies.head())
```

```python
user_movie_matrix = ratings.pivot_table(
    index='user_id', 
    columns='movie_id', 
    values='rating'
)

# Tratar dados faltantes: 
# Substituir NaN (não avaliado) por 0
user_movie_matrix = user_movie_matrix.fillna(0)

# Exibir a matriz processada
print("\n--- Matriz Usuário-Item (Formatada para K-Means) ---")
print(user_movie_matrix.head())

# Padronizar os dados para que o K-Means funcione melhor
# Isso ajuda o algoritmo a focar nos padrões de preferência
scaler = StandardScaler()
user_movie_matrix_scaled = scaler.fit_transform(user_movie_matrix)
```

```python
K = 10

kmeans_final = KMeans(
    n_clusters=K, 
    n_init=10, 
    random_state=42
)
kmeans_final.fit(user_movie_matrix_scaled)


# Adicionar os rótulos dos clusters, somando 1 (para começar em 1, não em 0)
user_movie_matrix['cluster'] = kmeans_final.labels_ + 1

print(f"\n--- Usuários agrupados em {K} clusters (1 a {K}) ---")
# Mostrar quantos usuários caíram em cada cluster (agora de 1 a 10)
print(user_movie_matrix['cluster'].value_counts().sort_index())
```

```python
K = 5

kmeans_final = KMeans(
    n_clusters=K, 
    n_init=10, 
    random_state=42
)
kmeans_final.fit(user_movie_matrix_scaled)

# Adicionar os rótulos dos clusters, somando 1 (para começar em 1, não em 0)
user_movie_matrix['cluster'] = kmeans_final.labels_ + 1

print(f"\n--- Usuários agrupados em {K} clusters (1 a {K}) ---")
# Mostrar quantos usuários caíram em cada cluster (agora de 1 a 5)
print(user_movie_matrix['cluster'].value_counts().sort_index())
```

```python

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

print("\n--- Gerando gráficos de visualização dos clusters ---")

# 1. Configurar o PCA
# Vamos reduzir as ~1682 dimensões (filmes) para apenas 2 (para o gráfico)
pca = PCA(n_components=2, random_state=42)

# 2. Aplicar o PCA nos mesmos dados que usamos para treinar o K-Means

data_2d = pca.fit_transform(user_movie_matrix_scaled)

# 3. Criar um novo DataFrame para o plot

plot_df = pd.DataFrame(
    data=data_2d, 
    columns=['PCA1', 'PCA2'],
    index=user_movie_matrix.index
)
plot_df['cluster'] = user_movie_matrix['cluster']


# --- GRÁFICO 1: USUÁRIOS ANTES DO AGRUPAMENTO ---
print("Plotando Gráfico 1 (Antes)...")
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1) # (1 linha, 2 colunas, gráfico 1)
plt.scatter(plot_df['PCA1'], plot_df['PCA2'], c='grey', alpha=0.6)
plt.title('Usuários Antes do Agrupamento (Visualização 2D)')
plt.xlabel('Componente Principal 1 (PCA1)')
plt.ylabel('Componente Principal 2 (PCA2)')


# --- GRÁFICO 2: USUÁRIOS DEPOIS DO AGRUPAMENTO (K=10) ---
print("Plotando Gráfico 2 (Depois)...")
plt.subplot(1, 2, 2) # (1 linha, 2 colunas, gráfico 2)
scatter = plt.scatter(
    plot_df['PCA1'], 
    plot_df['PCA2'], 
    c=plot_df['cluster'],  
    cmap='tab10',          
    alpha=0.6
)

# Título dinâmico (agora mostrará K=10)
plt.title(f'Usuários Agrupados por K-Means (K={K})')
plt.xlabel('Componente Principal 1 (PCA1)')
plt.ylabel('Componente Principal 2 (PCA2)')

# Legenda dinâmica (agora mostrará ticks de 1 a 10)
cbar = plt.colorbar(scatter, ticks=range(1, K + 1))
cbar.set_label('Cluster (Grupo de Gosto)')

plt.tight_layout()
plt.show()
print("Gráficos gerados.")
```

```python

def recomendar_para_grupo(cluster_id, n_recommendations=5):
    """
    Gera as 'Top Recomendações' para um cluster inteiro,
    baseado nos filmes que os membros do cluster mais 'amaram' (nota 5)
    E calcula o perfil de gênero do grupo.
    """
    
    print(f"\n--- Recomendações para o Grupo {cluster_id} ---")
    
    # 1. Pegar todos os usuários do cluster
    cluster_users = user_movie_matrix[
        user_movie_matrix['cluster'] == cluster_id
    ].index

    # 2. Encontrar filmes que os usuários do cluster gostaram muito (nota 5)
    cluster_ratings = ratings[
        (ratings['user_id'].isin(cluster_users)) & 
        (ratings['rating'] == 5)
    ]
    
    # 3. Contar quais filmes (ID) aparecem mais vezes com nota 5
    recommended_movie_ids = cluster_ratings['movie_id'].value_counts()

    # 4. Mesclar com títulos e mostrar
    final_recs = pd.merge(
        recommended_movie_ids.head(n_recommendations),
        movies,
        left_index=True, # O índice é o movie_id
        right_on='movie_id'
    )

    print(f"Top {n_recommendations} filmes preferidos por este grupo:")
    print(final_recs[['title', 'movie_id']])
    
    
    # 5. Pegar TODOS os movie_ids que o grupo deu nota 5
    all_liked_movie_ids = cluster_ratings['movie_id'].unique()
    
    # 6. Filtrar a tabela 'movies' para ter apenas esses filmes
    liked_movies_details = movies[movies['movie_id'].isin(all_liked_movie_ids)]
    
    # 7. Somar as colunas de gênero (usando 'genre_cols' da Célula 1)
    genre_profile = liked_movies_details[genre_cols].sum().sort_values(ascending=False)
    
    print("\nPerfil de Gênero do Grupo (Top 5 Gêneros):")
    # Filtra gêneros com 0 e exibe os 5 maiores
    print(genre_profile[genre_profile > 0].head(5))
    print("-" * 40)


# Gerar recomendações para todos os 10 grupos (de 1 a 10)
for i in range(1, K + 1):  # Loop de 1 até K (10)
    recomendar_para_grupo(cluster_id=i, n_recommendations=5)
```
